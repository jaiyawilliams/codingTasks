```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic.head()
```

```python
# Survival rate by gender
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.show()

# Survival rate by passenger class
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Survival rate by age group (creating age bins)
titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 30, 50, 80])
sns.barplot(x='age_group', y='survived', data=titanic)
plt.title('Survival Rate by Age Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# Survival rate by passenger class
sns.barplot(x='pclass', y='survived', hue='who', data=titanic)
plt.title('Survival Rate by Passenger Class and Person Type')
plt.show()
```

```python
# Survival rate by sex and age group
sns.barplot(x='who', y='survived', data=titanic)
plt.title('Survival Rate by Person Type')
plt.show()
```
